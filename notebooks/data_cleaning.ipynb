```python
import pandas as pd
import numpy as np
import warnings
from statistics import mode
warnings.filterwarnings("ignore")
```

## Holiday Package Prediciton

### 1) Problem statement.
"Trips & Travel.Com" company wants to enable and establish a viable business model to expand the customer base.
One of the ways to expand the customer base is to introduce a new offering of packages. Currently, there are 5 types of packages the company is offering * Basic, Standard, Deluxe, Super Deluxe, King. Looking at the data of the last year, we observed that 18% of the customers purchased the packages. However, the marketing cost was quite high because customers were contacted at random without looking at the available information.
The company is now planning to launch a new product i.e. Wellness Tourism Package. Wellness Tourism is defined as Travel that allows the traveler to maintain, enhance or kick-start a healthy lifestyle, and support or increase one's sense of well-being.
However, this time company wants to harness the available data of existing and potential customers to make the marketing expenditure more efficient.
### 2) Data Collection.
The Dataset is collected from https://www.kaggle.com/datasets/susant4learning/holiday-package-purchase-prediction
The data consists of 20 column and 4888 rows.

```python
df = pd.read_csv("D:\\Trips_and_Travel\\dataset\\Travel.csv")
```

```python
#Categorizing features
categorical_features = [col for col in df.columns if df[col].dtype == 'O']
numerical_features = [col for col in df.columns if df[col].dtype != 'O']
discrete_numerical = [col for col in numerical_features if df[col].nunique() <= 25]
continous_numerical = [col for col in numerical_features if df[col].nunique() >= 25]

print(f"Total number of features in our Dataset = {len(df.columns)}")
print(f"Number of Numerical Features in our Datset =  {len(numerical_features)}")
print(f"Total Number of discrete and Continous Numerical Features repectively =  {len(discrete_numerical)} , {len(continous_numerical)}")
print(f"Number of Categorical Fetaures = {len(categorical_features)}")
```

```python
# Number of Unique values in Categorical Features
unique_val = [df[uni].unique() for uni in categorical_features]

#Based on unique_val we make some transformation
df["Gender"].replace("Fe Male","Female",inplace=True)
df["MaritalStatus"].replace("Unmarried","Single",inplace=True)
```

```python
# Percentage of Null values 
Null_values_perc = [print(f"{col} ---- {round(df[col].isnull().mean()*100,2)}%") for col in df.columns if df[col].isnull().sum()>= 1]
```

```python
#filling missing values 

df["Age"].fillna(np.mean(df["Age"]),inplace=True)
df["TypeofContact"].fillna(mode(df["TypeofContact"]),inplace=True)
df["DurationOfPitch"].fillna(df["DurationOfPitch"].median(),inplace=True)
df["NumberOfFollowups"].fillna(mode(df["NumberOfFollowups"]),inplace=True)
df["PreferredPropertyStar"].fillna(mode(df["PreferredPropertyStar"]),inplace=True)
df["NumberOfTrips"].fillna(df["NumberOfTrips"].median(),inplace=True)
df["NumberOfChildrenVisiting"].fillna(df["NumberOfChildrenVisiting"].median(),inplace=True)
df["MonthlyIncome"].fillna(df["MonthlyIncome"].median(),inplace=True)
```

```python
# Dropping Unnecessary columns
df.drop("CustomerID",axis=1,inplace=True)
df["NumberOfPersonVisiting"]= df["NumberOfPersonVisiting"] + df["NumberOfChildrenVisiting"]
df.drop("NumberOfChildrenVisiting",axis=1,inplace=True)
```

```python
# Saving Cleaned Dataset

df.to_csv("Cleaned_data.csv")
```
